```python
#지역별 나이 분석
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_excel('./161.xlsx')

seven = 0
six = 0
four = 0
two = 0
one = 0
five = 0
three = 0
eight = 0
for i in range(0,len(df)):
    if df['AGE_1YEAR_UNIT'][i] >= 10 and df['AGE_1YEAR_UNIT'][i] < 20:
        one = one + 1
    if df['AGE_1YEAR_UNIT'][i] >= 20 and df['AGE_1YEAR_UNIT'][i] < 30:
        two = two + 1
    if df['AGE_1YEAR_UNIT'][i] >= 30 and df['AGE_1YEAR_UNIT'][i] < 40:
        three = three + 1
    if df['AGE_1YEAR_UNIT'][i] >= 40 and df['AGE_1YEAR_UNIT'][i] < 50:
        four = four + 1
    if df['AGE_1YEAR_UNIT'][i] >= 50 and df['AGE_1YEAR_UNIT'][i] < 60:
        five = five + 1
    if df['AGE_1YEAR_UNIT'][i] >= 60 and df['AGE_1YEAR_UNIT'][i] < 70:
        six = six + 1
    if df['AGE_1YEAR_UNIT'][i] >= 70 and df['AGE_1YEAR_UNIT'][i] < 80:
        seven = seven + 1
    if df['AGE_1YEAR_UNIT'][i] >= 80 and df['AGE_1YEAR_UNIT'][i] < 90:
        eight = eight + 1

age = [one, two, three, four, five, six, seven, eight]

x = np.arange(10, 90, 10)
generation = ['10','20', '30', '40', '50', '60', '70', '80']

plt.bar(x, age, color='y',width=1.0)

plt.xticks(x, generation)

plt.show()
```

```python
#지역별 성별 분석
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_excel('./157.xlsx')

male = 0
female = 0

for i in range(0,len(df)):
    if df['BSEX'][i] == 1:
        male =male + 1
    else:
        female = female +1

sex = [male, female]
x = np.arange(0,2)
xname = ['male','female']

plt.bar(x, sex, color='r',width=0.2)

plt.xticks(x, xname)

plt.show()
```

```python
#지역별 여행유형
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_excel('./157.xlsx')

four = 0
two = 0
one = 0
five = 0
three = 0

for i in range(0,len(df)):
    if df['AA3_1'][i] == 1:
        one = one + 1
    if df['AA3_2'][i] == 2:
        two = two + 1
    if df['AA3_3'][i] == 3:
        three = three + 1
    if df['AA3_8'][i] == 8:
        four = four + 1
    if df['AA3_9'][i] == 9:
        five = five + 1
        
category = [one, two, three, four, five]
x = np.arange(0,5)
xname = ['famous','food', 'activity', 'shopping', 'concert']

plt.bar(x, category, color='b',width=0.4)

plt.xticks(x, xname)

plt.show()
```

```python
#지역별 나이 분석
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_excel('./crawling_data/39.xlsx')

spring = 0
summer = 0
autumn = 0
winter = 0
for i in range(0,len(df)):
    if df['D_TRA1_SMONTH'][i] >= 3 and df['D_TRA1_SMONTH'][i] < 7:
        spring = spring + 1
    if df['D_TRA1_SMONTH'][i] >= 7 and df['D_TRA1_SMONTH'][i] < 10:
        summer = summer + 1
    if df['D_TRA1_SMONTH'][i] >= 9 and df['D_TRA1_SMONTH'][i] < 12:
        autumn = autumn + 1
    if df['D_TRA1_SMONTH'][i] == 12 or df['D_TRA1_SMONTH'][i] ==1 or df['D_TRA1_SMONTH'][i] ==2:
        winter = winter + 1
    
season = [spring, summer, autumn, winter]

x = np.arange(0,4)
xname = ['spring', 'summer', 'autumn', 'winter']

plt.bar(x, season, color='green',width=0.4)

plt.xticks(x, xname)

plt.show()
```
